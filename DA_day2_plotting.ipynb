plot the scores of 2 students in 5 different subjects.subjects as x-axis, marks as y-axis

```python
std1=[56,78,92,95,78]
std2=[77,89,91,23,45]
sub=["Maths","Science","English","Telugu","Social"]
plt.plot(sub,std1,color="r",label="student1")
plt.plot(sub,std2,color="m",label="student2")
plt.legend()
plt.title("Student Vs Marks")
plt.show()
```

```python
std1=[56,78,92,95,78]
std2=[77,89,91,23,45]
sub=["Maths","Science","English","Telugu","Social"]
plt.subplot(2,1,1)
plt.plot(sub,std1,color="r",label="student1")
plt.legend()
plt.subplot(2,1,2)#row,column,position
plt.plot(sub,std2,color="m",label="student2")
plt.legend()
plt.show()
```

create subplots[bar] showing the profits of 2 companies
Given: Revenew of Company A and B in Lakhs
plt.bar()
- Year      A                  B
- 2019      230               200
- 2020      560               160
- 2021      780               270
- 2022      127               127
- 2023      128               400

```python
years=np.array([2019,2020,2021,2022,2023])
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenews of company A",A)
print("Revenews of company B",B)
profA=np.diff(A)
print("Profits of A",profA)
profB=np.diff(B)
print("Profits of B",profB)
X=["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(X,profA,color='k',label="profit A")
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.bar(X,profB,color='g',label="profit B")
plt.legend(loc="best")
```
