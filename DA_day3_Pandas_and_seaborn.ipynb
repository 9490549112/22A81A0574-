To load dataset use:
 load_dataset("ds name")

- hue=different colour to different category
- style=different marker for diff category

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.catplot(data=tips,x="total_bill",y='day')
```

```python
dowjones=sns.load_dataset("dowjones")
dowjones.head()
```

```python
sns.relplot(data=dowjones,x="Date",y="Price",kind="line")
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

subjects=no of patients,  region = brain location

```python
sns.relplot(data=fmri,x='timepoint',y='signal',col='event')
```

```python
sns.catplot(data=fmri,x="timepoint",y='signal')
```

```python
sns.catplot(data=fmri,x='timepoint',y='signal',col='event',kind='violin',palette="bright")
```
